Data Types

Mean, Median, Mode





```python
import numpy as np
from scipy import stats

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

x = np.mean(speed)

print("Mean :", x)

x = np.median(speed)

print("Mod :", x)

x = stats.mode(speed)

print("Mode :", x)
```

Standard Deviaion, Variance and Percentile

```python
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]
err = np.std(speed)
print("Standard Deviation :",err)

var =  np.var(speed)

print("Varaince :",var)

x = np.percentile(speed, 75)

print("Percentile (75% Cars are below) :",x)
```

Regression

-Linear Regression

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))
print("Error in Perfection :", std_err)
plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

-Polynomial Regression

```python
from sklearn.metrics import r2_score
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel = np.poly1d(np.polyfit(x, y, 3))

myline = np.linspace(1, 22, 100)
print("The R square Result of the above programn :",r2_score(y, mymodel(x)))
plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()
```

-Multiple Regression

```python
import pandas
from sklearn import linear_model

df = pandas.read_csv("data.csv")

X = df[['Weight', 'Volume']]
y = df['CO2']

regr = linear_model.LinearRegression()
regr.fit(X, y)

#predict the CO2 emission of a car where the weight is 2300kg, and the volume is 1300cm3:
predictedCO2 = regr.predict([[2300, 1300]])

#dependencies on weight and volume as coefficents as {y = kx} giving us the value of k
print("Regression coefficient for CO2 emission :",regr.coef_)
print("Predicted CO2 emission :", predictedCO2)
predictedCO2 = regr.predict([[3300, 1300]])

print("Predicted CO2 emission after 1000kg rise is : ",predictedCO2)
```
